```python
# Q1.How many distinct (different) actors' last names are there?

query_1 = 'SELECT COUNT(DISTINCT(last_name)) AS "distinct_last_name" FROM sakila.actor'

data = pd.read_sql_query(query_1, engine)
data.head()
```

```python
# Q2.Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

query_2 = '''select *, dayofweek(rental_date),
case
when dayofweek(rental_date) in (2,3,4,5,6)
then "workday"
when dayofweek(rental_date) in (1,7)
then "weekend"
end as "day_type" from sakila.rental'''

data = pd.read_sql_query(query_2, engine)
data.head()
```

```python
# Q3.Get all films with ARMAGEDDON in the title.

query_3 = 'SELECT title FROM sakila.film WHERE title LIKE"%%ARMAGEDDON%%"'

data = pd.read_sql_query(query_3, engine)
data.head()
```

```python
# Q4.Get 10 the longest films.

query_4 = 'SELECT title, length FROM sakila.film ORDER BY length DESC LIMIT 10'

data = pd.read_sql_query(query_4, engine)
data.head()
```

```python
# Q5.How many films include Behind the Scenes content?

query_5 = '''SELECT DISTINCT (special_features),title FROM sakila.film
WHERE special_features LIKE "%%Behind the Scenes%%"''' 

data = pd.read_sql_query(query_5, engine)
data
```

```python
# Q6.Which kind of movies (rating) have a mean duration of more than two hours?

query_6 = 'SELECT rating, AVG(length) FROM sakila.film GROUP BY rating HAVING AVG(length) > 120'

data = pd.read_sql_query(query_6, engine)
data.head()
```

```python
# Q7 .Rank films by length (filter out the rows that have nulls or 0s in length column). 
#In your output, only select the columns title, length, and the rank.

query_7 = 'SELECT rating, AVG(length) FROM sakila.film GROUP BY rating HAVING AVG(length) > 120'

data = pd.read_sql_query(query_7, engine)
data.head()
```
